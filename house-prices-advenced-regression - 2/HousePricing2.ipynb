## Missing Value Handling

- Data hilang besar
- Split dua kategori (hilang > 1000 dan hilang < 1000)
- Bedakan penanganan berdasarkan kategori


```python
less = missing_values[missing_values < 1000 ].index           # dalam pandas => index dan series
over = missing_values[missing_values >= 1000].index
```

```python
# cek dengan out
print("data loss dengan miss value < 75%: ")
print(less)

print("\ndata over dengan miss value > 75%: ")
print(over)
```

```python
# HANDLING NUMERIC FITUR DATA < 74%
numeric_features = train[less].select_dtypes(include=['number']).columns
train[numeric_features] = train[numeric_features].fillna(train[numeric_features].median())       # isi dengan median
```

```python
# HANDLING KATEGORIK FITUR DATA > 75%
categorical_features = train[less].select_dtypes(include=['object']).columns
for column in categorical_features:
    train[column] = train[column].fillna(train[column].mode()[0])             # menggunakan looping karena mode mengembalikan ke series
    # mode()[0] -> jika modus lebih dari satu, maka ambil yang pertama
```

## Outliers Handling
- menggunana plt dan sns untuk visualisasikan outliers
- konsep IQR untuk bersihkan outlier

```python
# IMPORT LIBRARY UNTUK VISUALISASI
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

for feature in numeric_features:        # untuk setiap fiture dalam kategori fitur numerik
    plt.figure(figsize=(12, 8))         # buat canvas dengan lebar 12 tinggi 8
    sns.boxplot(x=train[feature])          # buat boxplot dengan nilai x dari fitur yang ada (akan ada 76 boxplot)
    plt.title(f"Boxplot of {feature}")  # berikan judul pada boxplot
    plt.show()                          # tunjukkan grafik
```

Ada dua pilihan untuk menangani: <br>
1. Hapus outlier
2. Mengganti dengan nilai moderat atau transformasi

Kali ini kita pakai hapus. Dengan metode IQR

```python
# Identifikasi outliers dg IQR Metode
Q1 = train[numeric_features].quantile(0.25)
Q3 = train[numeric_features].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Filter outlier dengan Mengamankan data di dalam IQR
condition = ~((train[numeric_features] < (Q1 - 1.5 * IQR)) | (train[numeric_features] > (Q3 + 1.5 * IQR))).any(axis=1)
train_filtered_numeric = train.loc[condition, numeric_features] # loc data yng sudh dikategorikan, data dari fitur numerik

# Menggabungkan kembali kolom numerik dengan kolom kategorikal
categorical_features = train.select_dtypes(include=['object']).columns
df = pd.concat([train_filtered_numeric, train.loc[condition, categorical_features]], axis=1)
```

# Data Splitting

```python
# pisahkan fitur dan target (Y)

X = df_lencoder.drop(columns=['SalePrice'])
y = df_lencoder['SalePrice']
```

```python
from sklearn.model_selection import train_test_split

# bagi dataset ke train dan test
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# menghitung panjang/jumlah data
print("Jumlah data: ",len(X))
# menghitung panjang/jumlah data pada x_test
print("Jumlah data latih: ",len(x_train))
# menghitung panjang/jumlah data pada x_test
print("Jumlah data test: ",len(x_test))
```

# Data Modelling

```python
# Melatih model 1 dengan algoritma LeasT Angle Regression

from sklearn import linear_model
lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)

# Melatih model 2 dengan algoritma Linear Regression
from sklearn.linear_model import LinearRegression
LR = LinearRegression().fit(x_train, y_train)

# Melatih model 3 dengan algoritma Gradient Boost Regressor
from sklearn.ensemble import GradientBoostingRegressor
GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(x_train, y_train)
```

# Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Evaluasi pada model LARS
pred_lars = lars.predict(x_test)
mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)

# Membuat dictionary untuk menyimpan hasil evaluasi
data = {
    'MAE': [mae_lars],
    'MSE': [mse_lars],
    'R2': [r2_lars]
}

# Konversi dictionary menjadi DataFrame
df_results = pd.DataFrame(data, index=['Lars'])
df_results
```

```python
# Evaluasi pada model Linear Regression
pred_LR = LR.predict(x_test)
mae_LR = mean_absolute_error(y_test, pred_LR)
mse_LR = mean_squared_error(y_test, pred_LR)
r2_LR = r2_score(y_test, pred_LR)

# Menambahkan hasil evaluasi LR ke DataFrame
df_results.loc['Linear Regression'] = [mae_LR, mse_LR, r2_LR]
df_results
```

```python
# Evaluasi pada model Gradient Boosting
pred_GBR = GBR.predict(x_test)
mae_GBR = mean_absolute_error(y_test, pred_GBR)
mse_GBR = mean_squared_error(y_test, pred_GBR)
r2_GBR = r2_score(y_test, pred_GBR)

# Menambahkan hasil evaluasi GBR ke DataFrame
df_results.loc['GradientBoostingRegressor'] = [mae_GBR, mse_GBR, r2_GBR]
df_results
```

Dari hasil model evaluation. Gradient Boosting memiliki nilai error yang paling kecil dengan evaluasi R2 mendekati 1 terbesar. Sehingga GradientBoosting adalah model yang dipilih
